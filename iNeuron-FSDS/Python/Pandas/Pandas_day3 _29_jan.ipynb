```python
#dictionary to dataframe
d2 = { 'key1' : [1,2,3,4,5,'Rakesh'] ,
      'key2' : [7,8,'Kumar',10,11,12] , 
      'key3' : [13,14,15,16,17,18],
      'key4' : [19,20,21,'iNeuron','pwskills',24]
}
```

```python
df2 = pd.DataFrame(d2)
```

```python
df2
```

```python
#merge operation - merging dataset based on certain condtion
pd.merge(df1,df2) #intersection of both the datasets
```

```python
#pd.merge() #by default merge() performs inner join
"""
pd.merge(
    left: 'DataFrame | Series',
    right: 'DataFrame | Series',
    how: 'str' = 'inner',
    on: 'IndexLabel | None' = None,
    left_on: 'IndexLabel | None' = None,
    right_on: 'IndexLabel | None' = None,
    left_index: 'bool' = False,
    right_index: 'bool' = False,
    sort: 'bool' = False,
    suffixes: 'Suffixes' = ('_x', '_y'),
    copy: 'bool' = True,
    indicator: 'bool' = False,
    validate: 'str | None' = None,
) -> 'DataFrame'
Parameters
----------
left : DataFrame
right : DataFrame or named Series
    Object to merge with.
how : {'left', 'right', 'outer', 'inner', 'cross'}, default 'inner'
    Type of merge to be performed.

    * left: use only keys from left frame, similar to a SQL left outer join;
      preserve key order.
    * right: use only keys from right frame, similar to a SQL right outer join;
      preserve key order.
    * outer: use union of keys from both frames, similar to a SQL full outer
      join; sort keys lexicographically.
    * inner: use intersection of keys from both frames, similar to a SQL inner
      join; preserve the order of the left keys.
    * cross: creates the cartesian product from both frames, preserves the order
      of the left keys.
"""
```

```python
pd.merge(df1,df2, how = 'inner' , on = 'key1')
```

```python
pd.merge(df1,df2, how = 'inner' , on = 'key4')
```

```python
pd.merge(df1,df2, how = 'left' , on = 'key4')#all data from left dataset(df1), and the the data common in both the datasets
```

```python
pd.merge(df1,df2, how = 'right' , on = 'key4')
```

```python
#outer merge => bringd, every combnation from left and right
pd.merge(df1,df2, how = 'outer' , on = 'key4')
```

```python
pd.merge(df1,df2, how = 'outer' , on = ['key4','key1'])
```
