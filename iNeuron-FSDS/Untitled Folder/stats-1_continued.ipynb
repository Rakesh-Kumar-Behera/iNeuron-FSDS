```python
#Normal distribution
import numpy as np
s = np.random.normal(0.5,0.2,1000)#normal(mean,std,data_points)
```

```python
#log normal ---> Normal Dist
sns.histplot(np.log(lgnd),kde = True)
```

### QQ Plot (Quantile-Quantile Plot)

when the quantiles of two variables are plotted against each other, then the plot obtained is known as quantile-quantile plot or qqplot. This plot provides summary of whether the distributions of two variables are similar or not w.r.t the locations.

```python
# QQ Plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab

def plot_data(sample):
    plt.figure(figsize = (10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde = True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist = 'norm',plot = pylab)
    plt.show()# QQ plot
```

```python
plot_data(data)
```

```python
plot_data(data_copy)
```

```python
plot_data(s)
```

```python
plot_data(lgnd)
```

```python
plot_data(np.log(lgnd))
```
